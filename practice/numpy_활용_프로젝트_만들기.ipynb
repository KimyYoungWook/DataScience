2. 궁금한 지역의 이름을 입력 받는다.


```python
import csv
f= open('age.csv',encoding='cp949')
data =csv.reader(f)
home = [] #입력 받은 지역의 데이터를 저장할 리스트
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
for row in data:
  if name in row[0] :
    for i in row[3:]:
      home.append(int(i.replace(',','')))
print(home)
```

3.궁금한 지역의 인구 구조를 저장한다.

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
f= open('age.csv',encoding='cp949')
data =csv.reader(f)
 #입력 받은 지역의 데이터를 저장할 리스트
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
for row in data:
  if name in row[0] :
    home =np.array(row[3:],dtype=int) #numpy 배열로 저장할때 데이터 타입 int로 변환
plt.figure(dpi=200)
plt.style.use('ggplot')# 격자 무늬 스타일로 지정 classic, dark_background
plt.rc('font', family = 'NanumBarunGothic')
plt.title(name+'지역의 인구 구조')
plt.plot(home)
plt.show()
```

4. 궁금한 지역의 인구 구조와 가장 비슷한 인구 구조를 가진 지역을 찾는다.


*   전국의 모든 지역 중 한곳 (B)을 선택 한다.
*   궁금한 지역 A의 0세 인구수에서 B의 0세 인구수를 뺀다
*   100세 이상 인구수에 해당 하는 값까지 반복한 후 각각의 차이를 모두 더한다.
*   전국의 모든 지역에 대해 반복하며, 그 차이가 가장 작은 지역을 찾는다.


궁금한 지역 A의 0세 인구 비율에서 B의 0세 인구 비율을 뺀다.


```python
import csv
import numpy as np
import matplotlib.pyplot as plt
f= open('age.csv',encoding='cp949')
data =csv.reader(f)
next(data)
data = list(data)
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
for row in data:
  row = [i.replace(',','') for i in row]
  if name in row[0] :
    home = np.array(row[3:],dtype = int )/int(row[2])
for row in data:
  row = [i.replace(',','') for i in row]
  away=np.array(row[3:],dtype=int)/int(row[2])
  
```

전국의 모든 지역에 대해 반복하여 그 차이가 가장 작은 지역을 찾고 시각화


```python
import csv
import numpy as np
import matplotlib.pyplot as plt
f= open('age.csv',encoding='cp949')
data =csv.reader(f)
next(data)
data = list(data)
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
mn=1 #최소값을 지정할 변수 생성 및 초기화
result_name = '' #최소값을 갖는 지역의 이름 저장할 변수 생성 및 초기화
result=0 # 최소값을 갖는 지역의 연령대별 인구 비율 저장할 변수 생성 및 초기화
for row in data:
  row = [i.replace(',','') for i in row]
  if name in row[0] :
    home = np.array(row[3:],dtype = int )/int(row[2])
for row in data:
  row = [i.replace(',','') for i in row]
  away=np.array(row[3:],dtype=int)/int(row[2])
  sum_h_a = np.sum(home-away)
  if sum_h_a <mn :
    mn = sum_h_a
    result_name = row[0]
    result= away
plt.figure(dpi=200)
plt.style.use('ggplot')# 격자 무늬 스타일로 지정 classic, dark_background
plt.rc('font', family = 'NanumBarunGothic')
plt.title(name+'지역의 인구 구조')
plt.plot(home,label= name)
plt.plot(away,label=result_name)
plt.legend()
plt.show()
```

우리가 찾고 있었던 것은 인구 구조가 가장 비슷한 지역 -> 즉 연령대별 인구 비율과 분포가 비슷해야 함<br>
두 지역의 차이가 합한 값이 0에 가까울 수록 인구 구조가 비슷 -> 음수 값이 선택됨<br>
거듭제곱 연산자를 통해 음수를 양수로 변환 처리 추가


```python
import csv
import numpy as np
import matplotlib.pyplot as plt
f= open('age.csv',encoding='cp949')
data =csv.reader(f)
next(data)
data = list(data)
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
mn=1 #최소값을 지정할 변수 생성 및 초기화
result_name = '' #최소값을 갖는 지역의 이름 저장할 변수 생성 및 초기화
result=0 # 최소값을 갖는 지역의 연령대별 인구 비율 저장할 변수 생성 및 초기화
for row in data:
  row = [i.replace(',','') for i in row]
  if name in row[0] :
    home = np.array(row[3:],dtype = int)/int(row[2])
for row in data:
  row = [i.replace(',','') for i in row]    
  away=np.array(row[3:],dtype=int)/int(row[2])  
  sum_h_a =np.sum((home-away)**2)
  if sum_h_a <mn and name not in row[0] :
    mn = sum_h_a
    result_name = row[0]
    result= away
plt.figure(dpi=200)
plt.style.use('ggplot')# 격자 무늬 스타일로 지정 classic, dark_background
plt.rc('font', family = 'NanumBarunGothic')
plt.title(name+'지역의 인구 구조')
plt.plot(home,label= name)
plt.plot(result,label=result_name)
plt.legend()
plt.show()
```
