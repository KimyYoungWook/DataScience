꺽은선 그래프로 표현하기

```python
import matplotlib.pyplot as plt
import csv
f= open('gender.csv',encoding='cp949')
data =csv.reader(f)
m=[]
f=[]
name=input('찾고싶은 지역의 이름을 알려주세요: ')
for row in data:
  if name in row[0]:
    for i in row[3:104]:# 남성데이터 : 3~103번 데이터 슬라이싱 - 정수 데이터로 변환 -m 리스트에 저장
      m.append(int(i.replace(",","")))
    for i in row[106:]:# 여성데이터 : 106번~ 끝까지 데이터 슬라이실 - 정수데이터로 변환 - f리스트에 저장
      f.append(int(i.replace(",","")))
    break
plt.figure(dpi=200)
plt.title(name+'지역의 남녀 성별 인구 분포')
plt.plot(m,label='남성')
plt.plot(f,label='여성')
plt.style.use('ggplot')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv
f= open('gender.csv',encoding='cp949')
data =csv.reader(f)
result=[] #남성인구 - 여성인구 값 담을 리스트 변수 선
name=input('찾고싶은 지역의 이름을 알려주세요: ')
for row in data:
  if name in row[0]:
    for i in range(3,104) :
      result.append(int(row[i].replace(',',''))-int(row[i+103].replace(',','')))
    break
plt.figure(dpi=200)
plt.title(name+'지역의 남녀 성별 인구 차이')
plt.rc('font', family='NanumBarunGothic') 
plt.bar(range(101),result)
plt.show()
```

위치,크기가 서로 다른 100개의 점 만들기

```python
import matplotlib.pyplot as plt
import random
x=[]
y=[]
size=[]

for i in range(100):
  x.append(random.randint(50,100))
  y.append(random.randint(50,100))
  size.append(random.randint(50,100))
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.scatter(x,y,s=size,cmap='jet',c=size,alpha=0.8)# alpha 투명도 추가 0~1: 0에 가까울수록 투명, 1에 가까울수록 불투명
plt.colorbar()
plt.show()
```

원의 크기는 인구수의 제곱근 추가

```python
import matplotlib.pyplot as plt
import csv
import math
f= open('gender.csv',encoding='cp949')
data =csv.reader(f)
m=[]
f=[]
size=[]
name=input('찾고싶은 지역의 이름을 알려주세요: ')
for row in data:
  if name in row[0]:
    for i in row[3:104]:# 남성데이터 : 3~103번 데이터 슬라이싱 - 정수 데이터로 변환 -m 리스트에 저장
      m.append(int(i.replace(",","")))
      size.append(math.sqrt(int(i.replace(",",""))))
    for i in row[106:]:# 여성데이터 : 106번~ 끝까지 데이터 슬라이실 - 정수데이터로 변환 - f리스트에 저장
      f.append(int(i.replace(",","")))
    break
plt.figure(dpi=200)
plt.title(name+'지역의 남녀 성별 인구 분포')
plt.style.use('ggplot')
plt.scatter(m,f,s=size,c=range(101),cmap='jet', alpha=0.5)
plt.plot(range(max(m)),range(max(m)),'g')
plt.show()
```
