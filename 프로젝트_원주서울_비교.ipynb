```python
import csv
import matplotlib.pyplot as plt
f = open("/content/Wonju.csv",encoding='cp949')
data=csv.reader(f)
next(data)
max=[]
avg=[]
min=[]
for row in data:
  if row[-1] != '':
   max.append(float(row[-1]))
   min.append(float(row[-2]))
   avg.append(float(row[2]))

plt.figure(dpi =200) #dpi 해상도 figsize 크기
plt.hist(max,bins=30,color='r',label='max')
plt.hist(min,bins=30,color='b',label='min')
plt.hist(avg,bins=30,color='g',label='avg')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f1 = open("/content/Wonju.csv",encoding='cp949')
data01=csv.reader(f1)
next(data01)
max_wonju=[]
max_seoul=[]
for row in data01:
  if row[-1] != '':
   max_wonju.append(float(row[-1]))
f1.close()
f2 = open("/content/Seoul.csv",encoding='cp949')
data02=csv.reader(f2)
next(data02)
for row in data02:
  if row[-1] != '':
   max_seoul.append(float(row[-1]))
f2.close()
plt.figure(dpi =200) #dpi 해상도 figsize 크기
plt.boxplot([max_wonju,max_seoul])
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f1 = open("/content/Wonju.csv",encoding='cp949')
data01=csv.reader(f1)
next(data01)
month=[]
for i in range(12):
  month.append([]) #
for row in data01:
  if row[-1] !='':
    month[int(row[0].split('-')[1])-1].append(float(row[-1]))

plt.figure(dpi=200)
plt.boxplot(month,showfliers=False)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f = open("/content/Wonju.csv",encoding='cp949')
data=csv.reader(f)
day=[]
next(data)
for i in range(31):
  day.append([]) # 

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '06':
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))
plt.figure(figsize=(10,5),dpi=200)
plt.boxplot(day,showfliers=False) # showfliers = > 이상치(아웃라이어) 삭제
plt.show()
```
