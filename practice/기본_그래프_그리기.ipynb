```python
import matplotlib.pyplot as plt
plt.plot([10,20,30,40]) # default가 y값임 자동으로 x값을 0부터 1씩 증가함
```

plt.plot(x,y)그래프 그리기

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[12,34,25,40]) # x,y 함수에 데이터 입력
plt.show() #그래프 보여주기
```

```python
import matplotlib.pyplot as plt
plt.title('plotting')#타이틀 속성 추가
plt.plot([1,2,3,4],[12,34,25,40]) # x,y 함수에 데이터 입력
plt.show() #그래프 보여주기
```

```python
import matplotlib.pyplot as plt
plt.title('plotting')#타이틀 속성 추가
plt.plot([10,20,30,40],label='asc') # 증가를 의미하는 asc 범례
plt.plot([40,30,20,10],label='desc') # 감소를 의미하는 desc 범례
plt.legend(loc=5) # 그래프에 범례 속성 추가
plt.show #그래프 보여주기
```

```python
import matplotlib.pyplot as plt
plt.title('color')#타이틀 속성 추가
plt.plot([10,20,30,40],color='skyblue',label='skyblue') # 하늘색 그래프
plt.plot([40,30,20,10],'pink',label='pink') # 분홍색 그래프
plt.legend(loc=5) # 그래프에 범례 속성 추가
plt.show #그래프 보여주기
```

```python
import matplotlib.pyplot as plt
plt.title('color')#타이틀 속성 추가
plt.plot([10,20,30,40],color='r',label='red') # 기본 색상은 약어로 가능빨간색 그래프
plt.plot([40,30,20,10],'g',label='green') # 초록색 그래프
plt.legend(loc=5) # 그래프에 범례 속성 추가
plt.show #그래프 보여주기
```

```python
import matplotlib.pyplot as plt
plt.title('linestyle')#타이틀 속성 추가
plt.plot([10,20,30,40],color='r',linestyle = '--',label='red') # --스타일 line
plt.plot([40,30,20,10],'g',ls = ':',label='green') # : 스타일 line -> ls로 대체 가능
plt.legend(loc=5) # 그래프에 범례 속성 추가
plt.show #그래프 보여주기
```

```python
import matplotlib.pyplot as plt
plt.title('marker')#타이틀 속성 추가
plt.plot([10,20,30,40],'r.',label='circle') #  점스타일 마크
plt.plot([40,30,20,10],'g^',label='triangle up') # 삼각형 스타일 마트
plt.legend(loc=5) # 그래프에 범례 속성 추가
plt.show #그래프 보여주기
```

```python
import matplotlib.pyplot as plt
plt.title('marker')#타이틀 속성 추가
# 'r,--' : <색상><마커모양><선모양> 코드로 작성 가능
plt.plot([10,20,30,40],'r.-',label='circle') #  점스타일 마크
plt.plot([40,30,20,10],'g^--',label='triangle up') # 삼각형 스타일 마트
plt.legend(loc=5) # 그래프에 범례 속성 추가
plt.show #그래프 보여주기
```

```python
import csv
import matplotlib.pyplot as plt
f = open('/content/Seoul.csv','r',encoding='cp949')
data =csv.reader(f)
header = next(data)
result = [] # 결측치를 제거한 최고 기온 리스트

for row in data : #최고 기온이 ''(빈문자열)이 아니면
  if row[-1] != '' :
    result.append(float(row[-1]))

plt.figure(dpi=200)
plt.plot(result,'r')
plt.show
```
