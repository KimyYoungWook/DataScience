```python
import matplotlib.pyplot as plt
import csv
f= open('gender.csv',encoding='cp949')
data =csv.reader(f)
m=[]
f=[]
name=input('찾고싶은 지역의 이름을 알려주세요: ')
for row in data:
  if name in row[0]:
    for i in row[3:104]:# 남성데이터 : 3~103번 데이터 슬라이싱 - 정수 데이터로 변환 -m 리스트에 저장
      m.append(-int(i.replace(",","")))
    for i in row[106:]:# 여성데이터 : 106번~ 끝까지 데이터 슬라이실 - 정수데이터로 변환 - f리스트에 저장
      f.append(int(i.replace(",","")))
    break
plt.figure(dpi=200)
plt.title(name+'지역의 남녀 성별 인구 분포')
plt.rc('font', family='NanumBarunGothic') 
plt.barh(range(101),m,label='남성')
plt.barh(range(101),f,label='여성')
plt.show()
```

제주도의 성멸인구 비율 표현하기

```python
import matplotlib.pyplot as plt
import csv
f= open('gender.csv',encoding='cp949')
data =csv.reader(f)
size=[]# 제주도 남성인구 합계, 여성인구합계
name=input('찾고싶은 지역의 이름을 알려주세요: ')
for row in data:
  if name in row[0]:
    m=0 # 남성 인구수 합계 변수
    f=0 # 여성 인구수 합계 변수
    for i in range(101):
      m = m + int(row[3+i].replace(',',''))
      f = f + int(row[106+i].replace(',',''))    
    break
size.append(f)
size.append(m)
plt.figure(dpi=200)
plt.title(name+'지역의 남녀 성별 인구 분포')
label=['남','여']
plt.pie(size,labels=label,autopct='%.1f',startangle=90)
plt.legend()
plt.show()
```
