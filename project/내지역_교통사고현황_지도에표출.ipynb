```python
import plotly.express as px
import pandas as pd
import folium
```

```python
# 데이터 전처리
df = pd.read_csv('도로교통공단_교통사고 정보.csv',encoding='euc-kr')
df=df.astype({'발생년월일시':'string' }) # 발생년월일시를 YYYY-MM-DD로 바꾸기 위해 string타입으로 변경 int형은 연속적이기때문에 처리하기힘듬
df['발생시간']=df.발생년월일시.str[8:] #발생년월일시에서 시를 분리하여 발생시간 열 생성
df=df.astype({'발생시간':'int64' }) #발생시간을 그래프에 표현하기 위해 숫자형으로 변경
df['발생년월일시']=pd.to_datetime(df['발생년월일시'].str[:8],
                            format='%Y-%m-%d',errors='raise') #발생년월일시를 datetime타입으로 변경
df_map=df[df.발생지시군구=='수원시'] 
map =folium.Map(location=[37.2635727,127.0286009])
#사망자수와 경상자수를 지도에 MArker 하기위해 float형식으로 변환
df_map=df_map.astype({'사망자수':'float64'})
df_map=df_map.astype({'부상자수':'float64'})
df_map=df_map.astype({'중상자수':'float64'})
df_map=df_map.astype({'경상자수':'float64'})
#데이터 시각화(지도에 Marker)
for n in df_map.index:
  cnt=df_map['사망자수'][n] + df_map['경상자수'][n] #사망자수 +경상자수  -> CircleMarker의 사이즈
  folium.CircleMarker([df_map['위도'][n],df_map['경도'][n]],
                       radius=(cnt*10),popup=df_map['사고유형'][n],color='#3186cc', fill_color='#3186cc').add_to(map)
map.save('수원시교통사고현황.html')
```

```python
df1=df[df.발생지시도=='경기']
fig=px.bar(df1,x='발생시간',y='사망자수',color ='발생지시군구',title='수원시 발생시간별/구별 교통사고 사망자 현황')
fig.show()
```
