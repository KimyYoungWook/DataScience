```python
#최고 기온 데이터만 리스트에 담아서 출력하기

import csv
import matplotlib.pyplot as plt
f = open("/content/Seoul.csv",encoding='cp949')
data=csv.reader(f)
next(data)
result=[]

for row in data :
  if row[-1] != '': #최고 데이터 값이 존재하면 , 결측치 제외
   result.append(float(row[-1]))

# 데이터  시각화
plt.figure(dpi=100)
plt.hist(result,bins=100,color='r')# 최고 기온  데이터를 빨간색 그래프로 그리기
plt.show()
```

1월부터 12월까지 최고기온 데이터 상자 그림으로 그리기

```python
import csv
import matplotlib.pyplot as plt
f = open("/content/Seoul.csv",encoding='cp949')
data=csv.reader(f)
next(data)
month = [[],[],[],[],[],[],[],[],[],[],[],[]] # 월별 데이터 저장할 month 변수 선언
for row in data:
  if row[-1] !='':
    month[int(row[0].split('-')[1])-1].append(float(row[-1]))# 월과 같은 번호의 인덱스에 월별 데이터 저장 (예: 1월 row[0].split('-')[1] -> month[0] row[0].split('-')[1]-1 , 2월 -> month[1])

plt.figure(dpi=200)
plt.boxplot(month)
plt.show()
```

8월 1일 부터 31일까지 최고 기온 데이터상자 그림으로 표현

```python
import csv
import matplotlib.pyplot as plt
f = open("/content/Wonju.csv",encoding='cp949')
data=csv.reader(f)
next(data)
max=[]
avg=[]
min=[]
for row in data:
  if row[-1] != '':
   max.append(float(row[-1]))
   min.append(float(row[-2]))
   avg.append(float(row[2]))

plt.figure(dpi =200) #dpi 해상도 figsize 크기
plt.hist(max,bins=50,color='r',label='max')
plt.hist(min,bins=50,color='b',label='min')
plt.hist(avg,bins=50,color='g',label='avg')
plt.legend()
plt.show()
```
