```python
import csv
f= open('age.csv',encoding='cp949')
data =csv.reader(f)
data_list=[row[3:] for row in data if '광교1동' in row[0]]
data_list_slicing_int=[int(i.replace(",","")) for i in data_list[0]]
print(data_list)
print(data_list_slicing_int)
```

```python
import matplotlib.pyplot as plt
plt.figure(dpi=200)
plt.style.use('ggplot')# 격자 무늬 스타일로 지정 classic, dark_background
plt.plot(data_list_slicing_int)
plt.show()
```
